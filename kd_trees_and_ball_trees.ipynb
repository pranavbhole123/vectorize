```python
import sys
import math
k=5 #number of dimension

class Node:
    def __init__(self,point) :
        self.point=point
        self.right = None
        self.left = None

        
```

```python
def newNode(point):
    return Node(point)


def search(root, point , depth):
    if not root:
        return False
    if root.point == point:
        return True
    
    #current dimension 
    cd = depth % k
    if point[cd] < root.point[cd]:
        return search(root.left, point , depth+1)
    
    else:
        return search(root.right,point,depth+1)
    

def insert(root , point , depth):
    if not root:
        return newNode(point)
    
    cd = depth%k

    if point[cd]<root.point[cd]:
        root.left = insert(root.left , point ,depth+1)

    else:
        root.right = insert(root.right , point , depth+1)

    return root

def insert1(root,point,depth):
    if search(root,point ,depth):
        print("already present")
        return root

    else:
       return  insert(root,point,depth)
    

def distance_square(point1,point2):
    sum=0
    for i in range(k):
        sum+= (point1[i]-point2[i]) * (point1[i]-point2[i])
    return sum

best_dist=999999999999999999
best=[]
def nearest(root , point ,depth , best , best_dist):

    if (not root) :
        return best ,best_dist
    
    dist = distance_square(point , root.point)
    if dist<best_dist:
        best = root.point
        best_dist = dist

    cd = depth%k

    if point[cd] <root.point[cd]:
        best,best_dist=nearest(root.left , point , depth+1,best,best_dist)
        #nearest(root.right , point , depth+1, rect.trimeright(cd ,root.point))
    else:
        best , best_dist=nearest(root.right , point , depth+1, best , best_dist)
        #nearest(root.left , point , depth+1, rect.trimeleft(cd ,root.point))
        

    #check if other side is important
    if(abs(point[cd]-root.point[cd])<math.sqrt(best_dist)):
        if point[cd]<root.point[cd]:
            best,best_dist = nearest(root.right,point,depth+1,best,best_dist)
        else:
            best,best_dist = nearest(root.left,point,depth+1,best,best_dist)
    return best ,best_dist


```

```python
def farthest(points):
    max = -1
    p1, p2 = None, None
    n=len(points)
    for i in range(n):
        for j in range(n):
            dist=distance_square(points[i],points[j])
            if dist>max:
                p1=points[i]
                p2 = points[j]
                max = dist
    return p1 ,p2, max

  
```

```python
def partition(points,p1,p2):
    n=len(points)
    l=[]
    m=[]
    
    for i in range(n):
        
        #print(points[i])
        if distance_square(points[i],p1)<distance_square(points[i],p2):
            l.append(points[i])
        else:
            m.append(points[i])
    return l ,m
    
depth_of_ball_trees=10
```

```python
l,m = partition(points,[9,1],[10,19])
print(l,m)
```

```python
def build_ball(points, level, center):
    if not points or level >= depth_of_ball_trees:
        return None

    p1, p2, diameter = farthest(points)
    #print(p1,p2)
    #median = [(x + y) / 2 for x, y in zip(p1, p2)]
    
    #print(median)
    root = newBall(points, diameter / 4, center)

    l, m = partition(points, p1, p2)
    
    # Update the root's left and right children
    root.right = build_ball(m, level + 1,p2)
    root.left = build_ball(l, level,p1)

    return root

def print_ball_tree(root, depth=0):
    if root is not None:
        print("  " * depth + f"Center: {root.centre}, Radius: {root.radius}, data : {root.data}")
        print_ball_tree(root.left, depth + 1)
        print_ball_tree(root.right, depth + 1)
    


```

```python
current_best=[]
best_distance=9999999999999


def nearest_neb(root,point):
    global best_distance , current_best
    if not root:
        return current_best , best_distance
    
    if root.centre:
        dist=distance_square(point,root.centre)
    
        if dist< best_distance:
            current_best=root.centre
            best_distance=dist
    
    if not root.right and root.left:
        nearest_neb(root.left,point)
    elif not root.left and root.right:
        nearest_neb(root.right,point)

    if root.right and root.left :
        if distance_square(point,root.left.centre)<distance_square(point,root.right.centre):
            side="left"
        else:
            side="right"
        closer_child, farther_child = (root.left, root.right) if side == "left" else (root.right, root.left)
        current_best,best_distance = nearest_neb(closer_child,point)

        if root.centre:
            if (best_distance)>distance_square(point,root.centre):
                current_best,best_distance = nearest_neb(farther_child,point)
    
    return current_best , best_distance
    
```
