```python
import numpy as np
import random
from numpy.linalg import norm
from numpy import array

#global variables
dim =6
length =1000000
p = 10 # number of hash tables to be geneerated
```

```python
def genhyp():
    v = []
    for i in range(dim):
        v.append(np.random.rand() * 2 - 1)  
    return v

def dotprod(v , u):
    sum=0
    for i in range(dim):
        sum += v[i]*u[i]
    return sum

def hashp(points,k):
    n = length
    dicty={}
    v=[]
    for i in range(k):
        v.append(genhyp())

    for i in range(n):
        s=""
        for j in range(k):
            p = dotprod(points[i],v[j])
            if(p>=0):
                s+="1"
            else:
                s+="0"
        if s not in dicty:
            dicty[s]=[]
        dicty[s].append(points[i])
    return dicty,v

def print_keys_and_lengths(hash_dict):
    for key, value_list in hash_dict.items():
        print("Key:", key)
        print("Length of values:",len(value_list) )
        print()
```

```python


def generate_matrix(k):
    # Create a matrix with 1000 rows and k columns
    matrix = [[np.random.uniform(-1, 1) for _ in range(k)] for _ in range(10000000)]
    return matrix


# Example: generate a matrix with 5 columns
result_matrix = generate_matrix(dim)


res,planes=hashp(result_matrix,6) # here 4 is number of hyperplanes nee
print_keys_and_lengths(res)
```

```python


y=[] # list of all dictionaries
z=[] # list of all planes
for i in range(p):
    temp1,temp2 = hashp(result_matrix,6)
    y.append(temp1)
    z.append(temp2)

# niw to search for query point we will have to search through

def search(dic_arr, plan,query,k): # k is the number of hyperplanes
    count=0
    maxy=-1
    res=None
    for i in range(p):
        
        s=""
        for j in range(k):
            
            temp = dotprod(query,plan[i][j])
            if(temp>=0):
                s+="1"
            else:
                s+="0"
        

        if s in dic_arr[i]:       
            for items in dic_arr[i][s]:
                count+=1
                cosine = np.dot(query,items)/(norm(items)*norm(query))
                if cosine>maxy:
                    res = items
                    maxy=cosine

    return maxy,res,count

            

```
